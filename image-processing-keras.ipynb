```python
#library
from tensorflow.keras.preprocessing.image import load_img
#eğer keras hata verirse başına tensorflow ekle 
#from tensorflow.keras.preprocessing.image import 
```

```python
#reading - loading image
img = load_img("datasets/yoda.jpg")
print("Type:" , type(img))
print("Format:" , img.format)
print("Size:" , img.size)
print("Mode:" , img.mode)
```

```python
#displaying image
#keras'ta yüklenenen görüntüleri okuyabilmek için matplotlib kütüphanesi kullanılır.
#pip install matplotlib

import matplotlib.pyplot as plt
img = load_img("datasets/yoda.jpg")
img1=load_img("datasets/clone.jpg")
plt.imshow(img)
plt.show()
#imshow ve show komutlarını alt alta birlikte kullandığımız zaman görüntünün çıktısını verir.
plt.imshow(img1)
plt.show()

```

```python
#PIL --> numpy array
#görüntüyü bir modele dahil edebilmek için array formatına dönüştürmek gerekir. 
from tensorflow.keras.preprocessing.image import load_img, img_to_array, array_to_img

img=load_img("datasets/clone.jpg")
print("type:" , type(img))

img = img_to_array(img)
print("Type:" , type(img))

#numpy array --> PIL 
img= array_to_img(img)
print("Type:", type(img))
```

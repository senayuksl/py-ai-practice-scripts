ResNet50

```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
import numpy as np
```

```python
model = ResNet50(weights="imagenet")
```

```python
test_path = "elephant.jpg"
img= image.load_img(test_path, target_size=(224,224)) #target_size, yeniden boyutlandırmadır. resnet boyutları 224x224 old için ona uygun boyutlandırıldı
img= image.img_to_array(img)
img= np.expand_dims(img, axis=0) #4 boyutlu veriye çevirir
img= preprocess_input(img)

prediction = model.predict(img)

```

```python
prediction
```

```python
prediction = decode_predictions(prediction, top=3)[0]
prediction
```

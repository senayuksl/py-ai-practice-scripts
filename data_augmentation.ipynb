```python
#librart
from tensorflow.keras.preprocessing.image import load_img, img_to_array, ImageDataGenerator
```

```python

img= load_img("datasets/star/yoda.jpg")
type(img)
"""
1. (100,100,3) normal bir görüntü bu şekildedir. görüntüyü batch'e dönüştürmemiz gerekir. 
2. PIL -> array
3. (veri sayısı, 100, 100, 3) batch'e dönüştürülür.yeni oluşsn çıktı yandaki gibidir.
"""
img_array = img_to_array(img)
type(img_array)
print("shaped:" , img_array.shape)

reshaped_img_array = img_array.reshape((1, ) + img_array.shape)
type(reshaped_img_array)
print("reshaped:" , reshaped_img_array.shape)
```

```python
data_generator= ImageDataGenerator(rotation_range=65, 
                                   width_shift_range=0.2, 
                                   height_shift_range=0.2,
                                   zoom_range=0.2,
                                   horizontal_flip=True, 
                                   fill_mode="nearest" )
```

```python
counter=0
for i in data_generator.flow(reshaped_img_array, batch_size=1, 
                              save_to_dir="datasets/star/augmented_yoda", #hata almamak için tırnak içinde ve eğik çizgiyle yazılır.
                              save_prefix="jedi",
                              save_format="jpeg"):
    
    counter = counter+1
    print("info", counter)
    if counter ==100: #resmi kaç kere tekrarlayıp çoğaltacağı
        break
        
    
```

```python
data_generator_directory= ImageDataGenerator(rotation_range=65, 
                                   width_shift_range=0.2, 
                                   height_shift_range=0.2,
                                   zoom_range=0.2,
                                   horizontal_flip=True, 
                                   vertical_flip=True,          
                                   fill_mode="nearest" )
```

```python
counter_d=0
for i in data_generator_directory.flow_from_directory("datasets/star/yoda_clone", batch_size=1, 
                              save_to_dir="datasets/star/augmented_yoda_clone", #hata almamak için tırnak içinde ve eğik çizgiyle yazılır.
                              save_prefix="jedi",
                              save_format="jpeg"):
    
    counter_d = counter_d+1
    print("info", counter_d)
    if counter_d ==100: #resmi kaç kere tekrarlayıp çoğaltacağı
        print("done!")
        break
        
```
